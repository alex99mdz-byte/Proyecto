```python
import pandas as pd
import matplotlib.pyplot as plt
 
 # %%
#declaracion de datos a utilizar en vectores y matrices
nombres=['Ana', 'Luis', 'Carlos', 'Marta']
calificaciones=[[5,5.5,10], [9,8.7,10], [3,5,7], [10,10,10]]
 
# %%
#crear data frame
df=pd.DataFrame({'nombre':nombres,'calificaciones':calificaciones})
df
 
# %%
#calculos
medias=[]
for x in df['calificaciones']:
    medias.append(sum(x)/len(x))
   
df['media']=medias
cnt=0
   
 
# %%
df
 
# %%
#grafica
plt.bar(df['nombre'],df['media'],color='blue')
plt.xlabel('Estudiantes')
plt.ylabel('Promedios')
plt.title('Promedios de calificaciones por estudiante')
for i in df['media']:
    plt.text(cnt, i + 0.1, str(round(i,2)), ha='center')
    cnt+=1
plt.show()
```
